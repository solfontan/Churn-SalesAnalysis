```python
import pandas as pd
from functools import reduce

# Lista de archivos CSV
file_paths = [
    r"data\clean\customer_info_clean.csv",
    r"data\clean\location_data_clean.csv",
    r"data\clean\online_options_clean.csv",
    r"data\clean\online_services_clean.csv",
    r"data\clean\payment_method_clean.csv",
    r"data\clean\status_analysis_clean.csv"
]

# Cargar los DataFrames
dfs = [pd.read_csv(file) for file in file_paths]

# Unir todos los DataFrames en 'customer_id'
df_clean = reduce(lambda left, right: pd.merge(left, right, on="customer_id"), dfs)

# Mostrar las primeras filas
df_clean.head()
```
